```python
# Importar librerías necesarias
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler

# Cargar dataset
digits = load_digits()
X = digits.data
y = digits.target

# Mostrar algunas imágenes del dataset digits
plt.figure(figsize=(12, 6))
for i in range(10):  # Mostrar 10 imágenes
    plt.subplot(2, 5, i + 1)
    plt.imshow(digits.images[i], cmap='binary', interpolation='nearest')
    plt.title(f"Label: {digits.target[i]}")
    plt.axis('off')
plt.suptitle("Ejemplos de Imágenes del Dataset Digits", y=1.05)
plt.tight_layout()
plt.show()



# Escalar los datos
X_scaled = StandardScaler().fit_transform(X)

# Reducción de dimensionalidad con PCA para visualización inicial
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Visualización con PCA
plt.figure(figsize=(10, 8))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, palette='viridis', legend='full')
plt.title("PCA de Digits Dataset")
plt.xlabel("Componente PCA 1")
plt.ylabel("Componente PCA 2")
plt.legend(title='Dígitos')
plt.show()

# Aplicar t-SNE
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

# Visualización con t-SNE
plt.figure(figsize=(10, 8))
sns.scatterplot(x=X_tsne[:, 0], y=X_tsne[:, 1], hue=y, palette='viridis', legend='full')
plt.title("t-SNE de Digits Dataset")
plt.xlabel("Componente t-SNE 1")
plt.ylabel("Componente t-SNE 2")
plt.legend(title='Dígitos')
plt.show()
```
